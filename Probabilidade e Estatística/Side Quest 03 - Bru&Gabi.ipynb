O andar do bêbado
=================



## Objetivo



Um senhor que tomou uma ou duas doses a mais do que deveria está andando por um campo plano, infinito e livre de obstáculos. A cada minuto, este senhor dá um passo de um metro para norte, sul, leste ou oeste. É sempre um passo por minuto e sempre em apenas uma direção (todas as direções têm a mesma chance de serem escolhidas).

Faça gráficos mostrando:

1.  A trajetória do senhor andante no plano;

2.  A posição em $x$ do senhor andante em função do tempo (plote uma linha indicando o valor médio);

3.  A posição em $y$ do senhor andante em função do tempo (plote uma linha indicando o valor médio);

4.  A distância relativa percorrida pelo senhor andante em função do tempo (a distância relativa é a distância Euclidiana entre a posição no tempo atual e a posição no tempo zero).

Considere que o senhor andante inicia sua trajetória no tempo $t=0$ na coordenada $(0,0)$. Considere um tempo de pelo menos 1000 minutos para plotar seus gráficos. Discuta sobre os resultados obtidos.



## Imports e funçõe

<p style='text-align:justify'> Para essa atividade, usamos o módulo <code>numpy</code> para calcular a distância entre os pontos na função <code>distancia_origem</code>, o <code>statistics</code> para computar a média das coordenadas em $x$, o <code>pyplot</code> para plotar os gráficos e, por fim, a função <code>choice</code> do módulo <code>random</code>.

```python
import numpy as np
import statistics as st
import matplotlib.pyplot as plt
from random import choice


def distancia_origem(array):
    """Calcula a distância de pontos até a origem
    
    A partir de um array bidimensional, a função calcula a distância 
    euclidiana até a origem.
    
    Args:
        array: array 2D que corresponde às coordenadas x e y.
    
    Returns:
        A distância euclidiana até a origem dos pontos x e y.
    
    """
    distancia_ao_quadrado = array[0]** 2 + array[1]**2
    distancia = np.sqrt(distancia_ao_quadrado)
    return(distancia)


def andar():
    """Sorteia a próxima direção da trajetória e adiciona à posição.
    
    Essa função sorteia uma direção entre norte, sul, leste e oeste e adiciona
    essa nova direção à posição.
    
    """
    direcao = random.choice(lista_sorteio)
    Posicao[0] += direcao[0]
    Posicao[1] += direcao[1]
```

## Trajetória do senhor andante no plano

<p style='text-align:justify'> Para calcular a trajetória do senhor andante, criamos a variável <code>Posicao</code> para abrigar a posição de origem. Além disso, também criamos variáveis para as direções <code>Norte</code>,<code>Sul</code>,<code>Leste</code> e <code>Oeste</code>. Com isso, criamos uma lista<code>lista_sorteio</code> que abrigue as quatro possíveis direções e também uma outra lista <code>lista_de_posicoes</code> vazia. Assim, computamos o número de passos que o senhor iria andar e iteramos um loop com um range do número de passos. No loop. adicionamos a origem à <code>lista_de_posicoes</code> e executamos a funçaõ <code>andar()</code> para atualizar a nova direção. Dessa forma, criamos a lista <code>x_coords</code> para as coordenadas $x$ que foram iteradas e <code>y_coords</code> para a as coordenadas $y$. Por fim, usamos o <code>pyplot</code> para plotar os gráficos.

```python
Posicao = [0, 0]

Norte = [0, 1]
Sul = [0, -1]
Leste = [1, 0]
Oeste = [-1, 0]

lista_sorteio = [Norte, Sul, Leste, Oeste]

lista_de_posicoes = []



Numero_de_passos = 1000

for passo in range(Numero_de_passos):
    lista_de_posicoes.append(list(Posicao))
    andar()

x_coords = [posicao[0] for posicao in lista_de_posicoes]
y_coords = [posicao[1] for posicao in lista_de_posicoes]


plt.figure(figsize=(8, 8))
plt.plot(x_coords, y_coords, marker='o', linestyle='-', color='#00BFFF')
plt.title('Trajeto do Movimento')
plt.ylabel('Sul-Norte')
plt.xlabel('Oeste-Leste')
plt.xticks(range(min(x_coords), max(x_coords)+1,3))
plt.yticks(range(min(y_coords), max(y_coords)+1,3))
plt.grid(True)

plt.show()
```

## A posição em $x$ do senhor andante em função do tempo 

<p style='text-align:justify'> Para plotar a posição em $x$ em função do tempo, criamos uma lista de minutos usando <code>*range(1000)</code>, partindo de $t=0$ até $t=1000$. Após isso, usamos o <code>pyplot</code> novamente para plotar um gráfico em que as coordenadas do eixo $x$ estão no eixo vertical e o tempo no eixo horizontal.

```python
minutos = [*range(1000)]


plt.figure(figsize=(8, 8))
plt.plot(minutos, x_coords, marker='o', linestyle='-', color='#004953')
plt.axhline(st.mean(x_coords), linestyle="--", color="red", label="Média")
plt.title('Posição de x do senhor andante em função do tempo')
plt.ylabel('Posição em X')
plt.xlabel('Tempo (minutos)')
plt.grid(True)
```

## A posição em $y$ do senhor andante em função do tempo 

<p style='text-align:justify'> Aqui, fizemos a mesma coisa que no código passado, apenas alterando a variável de coordenadas $x$ para $y$, deixando ele no eixo vertical também.

```python
minutos = [*range(1000)]


plt.figure(figsize=(8, 8))
plt.plot(minutos, y_coords, marker='o', linestyle='-', color='#FFD700')
plt.title('Posição de y do senhor andante em função do tempo')
plt.axhline(st.mean(y_coords), linestyle="--", color="red", label="Média")
plt.ylabel('Posição em Y')
plt.xlabel('Tempo (minutos)')
plt.grid(True)
```

## A distância relativa percorrida pelo senhor andante em função do tempo 

<p style='text-align:justify'>Por fim, para caclcular a distância relativa percorrida em função do tempo, criamos a lista <code>lista_distancias</code> para armazenar as distâncias e criamos um loop iterando a lista <code>lista_de_posicoes</code> com as posições. Com o loop, rodamos a função <code>distancia_origem()</code> para adicionar as distâncias euclidianas dos pontos na lista <code>lista_distancias</code> e usamos ele como variável para plotar um gráfico em função do tempo.

```python
lista_distancias = []

for posicao in lista_de_posicoes:
    distancia = distancia_origem(posicao)
    lista_distancias.append(distancia)
    
plt.figure(figsize=(8, 8))
plt.plot(minutos, lista_distancias, marker='o', linestyle='-', color='#40E0D0')
plt.title('A distância relativa percorrida pelo senhor andante em função do tempo')
plt.ylabel('Distancia')
plt.xlabel('Tempo (minutos)')
plt.grid(True)
```
